Applying Functions

Creating dataFrame from List

```
names = ['John', 'Anna', 'Peter','Ronak', 'Megha']
ages = [25, 30, 35, 99, 14]
cities = ['New York', 'Paris', 'London', 'Delhi', 'Mumbai']
```

```
import pandas as pd
df_new = pd.DataFrame(list(zip(names, ages, cities)), columns=['Name', 'Age', 'City'])
```

```
df_new.head()
```

Creating dataFrame from dictionary

```
data = {'Name': ['John', 'Anna', 'Peter'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Paris', 'London']}
```

```
df_dict = pd.DataFrame(data)
```

```
df_dict.head()
```

****************************************************************************************************************************************************************

```
my_data = {
    'Name': ['John', 'Anna', 'Peter', 'Emily', 'Michael'],
    'Age': [18, 17, 18, 17, 16],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Exam_Score': [85, 90, 75, 80, 70],
    'Study_Hours': [5, 6, 4, 5, 3],
    'Grade': ['A', 'A', 'B', 'B', 'C']
}
```

```
my_df = pd.DataFrame(my_data)
my_df.head()
```

Bar plot using pandas

```
my_df.plot.bar(x='Name',y= "Exam_Score")
```

scatter plot using DataFrame’s values as coordinates

```
my_df.plot.scatter(x='Study_Hours', y='Exam_Score')
```

line plot using DataFrame’s values as coordinates

```
my_df.plot.line(x='Study_Hours', y='Exam_Score')
```
