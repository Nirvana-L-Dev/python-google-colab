```
# importing library
import matplotlib.pyplot as plt
import numpy as np
```

###Line Plot

```
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot
plt.plot(x, y)
plt.xlabel('Names')
plt.ylabel('Exam_Scores')
plt.title('This is the title for a plot')
plt.show()
```

###Scatter Plot

```
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

###Bar Plot

```
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plot
plt.barh(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

###Histogram

```
data = np.random.randn(1000)

# Plot
plt.hist(data, bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

###Pie Chart

```
# Data
categories = ['Rent', 'Groceries', 'Utilities', 'Transportation', 'Entertainment']
expenses = [1000, 500, 300, 200, 150]  # Example expenses in dollars

# Plot
plt.pie(expenses, labels=categories, autopct='%1.1f%%')
plt.title('Monthly Household Expenses')
plt.show()
```

###Subplots

```
# Data
categories = ['Rent', 'Groceries', 'Utilities', 'Transportation', 'Entertainment']
expenses = [1000, 500, 300, 200, 150]  # Example expenses in dollars

# Plot
plt.figure(figsize=(10, 10))

# Display percentages with one decimal place
plt.subplot(1, 2, 1)
plt.pie(expenses, labels=categories, autopct='%1.1f%%')
plt.title('With 1 Decimal Place')

# Display percentages with two decimal places
plt.subplot(1, 2, 2)
plt.pie(expenses, labels=categories, autopct='%1.2f%%')
plt.title('With 2 Decimal Places')
plt.show()
```

###Customizing Plots

```
# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
plt.plot(x, y, label='sin(x)', color='green', linestyle='--', linewidth=2, markersize=5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Plot')
plt.legend()
plt.grid(True)
plt.show()
```

###Saving Plots

```
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.savefig('plot.png')
plt.show()
```
