# Indexing and Slicing On Strings

```python
name = "Siri"
print(name[0])
print(name[1])
print(name[2])
print(name[3])
```

```python
# Example string
s = "Hello, world!"

# Slicing
print(s[1:5])
print(s[7:])
print(s[:5])
print(s[-1])
```

# String Methods

```python
text = "apple, banana, cherry"
words = text.split(", ")
print(words)

words_joined = ", ".join(words)
print(words_joined)
```

```python
phrase = "PythON ProgrAmMing"
print(phrase.lower())

print(phrase.upper())

```

```python
text = "hello world"
print(text.capitalize())

print(text.index("o"))

print(text.count("l"))
```

```python
statement = "Hello World"
print(statement.isupper())

print(statement.islower())

print(statement.startswith("Hello"))

print(statement.endswith("world"))
```

```python
text = "hello world"

first_o_index = text.index("o")
print("Index of first 'o':", first_o_index)

second_o_index = text.index("o", first_o_index + 1)
print("Index of second 'o':", second_o_index)
```
