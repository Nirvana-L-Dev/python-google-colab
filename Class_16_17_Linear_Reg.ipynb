# **Importing Required Libraries/Modules**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

# **Data Collection**

```python
import pandas as pd

# Create a dictionary with data
data = {
    'Hours Of Study': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4],
    'Marks': [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69]
}

# Create a DataFrame
df = pd.DataFrame(data)
df
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Hours Of Study', y='Marks', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

# Save the DataFrame to a CSV file
df.to_csv('hours_marks.csv', index=False)
```

# **Data Exploration**

```python
# Load dataset
data = pd.read_csv('hours_marks.csv')
data.head()
```

```python

# Features and target variable
X = df[['Hours Of Study']]
y = df['Marks']
X
y
```

# **Data Splitting**

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#100% - 80% train, 20%test
# 1 - 0.8, 0.2
```

# **Model Building**

```python
# from sklearn.linear_model import LinearRegression
# Initialize the model
model = LinearRegression()
```

# **Model Training**

```python

# Train the model
model.fit(X_train, y_train)
```

```python

# Make predictions
y_pred = model.predict(X_test)
```

```python

# Plotting the regression line and data points
plt.scatter(X, y, color='blue')  # Data points
plt.plot(X, model.predict(X), color='red')  # Regression line
plt.xlabel('Hours of Study')
plt.ylabel('Marks')
plt.title('Linear Regression - Hours of Study vs Marks')
plt.show()
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
