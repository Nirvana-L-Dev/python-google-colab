# **Tuple**

```python
# Creating a tuple
example_tuple = (1, 2, 3, 4, 2, 2)

# Indexing
print("Element at index 0:", example_tuple[0])
```

```python

#slicing
sliced_tuple = example_tuple[1:4]  # Gets elements from index 1 to 3
print("Sliced Tuple:", sliced_tuple)
```

```python
# Count method
print("Count of 2 in Tuple:", example_tuple.count(2))
```

```python
example_tuple = (1, 7, 3, 4, 2, 2)
# Index method
print("First index of 2:", example_tuple.index(4))
```

# **Set**



1.   **Sets are unordered collections of unique elements in Python.**
2.   **Duplicate Items are Not allowed**



```python
# Creating sets
set_a = {1, 2, 3, 4, 30 , 40}
set_b = {80,90}

#super set and sub set

# issuperset and isdisjoint methods
print("A is superset of B:", set_a.issuperset(set_b))
print("A is disjoint with B:", set_a.isdisjoint(set_b))
```

```python

# update method
set_a.update(set_b)
print("Updated Set A:", set_a)
```

```python

# add, remove, discard, pop, clear, copy methods
set_a.add(60)  # Adding element to set
print(set_a)
set_a.remove(2)  # Removing element throws error if element not present
print(set_a)
```

```python
set_a.discard(80)  # Discards element if present, no error if not
print(set_a)
```

```python
print(set_a)
popped_element = set_a.pop()  # Removes an arbitrary element
print("Popped Element:", popped_element)
print(set_a)
```

```python

seta={3,4,5}
set1 = seta.copy()  # Creates a copy of the set
print(set1)
```

# **Dictionary** key-value pairs

```python
# Creating a dictionary
my_dict = {'name': 'Alice', 'age': 25, 'location': 'Wonderland'}

# Using get, clear, copy, update, keys, values, and items methods
print(my_dict.get('age'))  # Safer way to access values
```

```python
print(my_dict.keys())
print(my_dict.values())
print(my_dict.items())
```

```python

# Update method
my_dict.update({'age': 40, 'email': 'alice@example.com'})
print("Updated Dictionary:", my_dict)
```

```python

# pop, popitem, fromkeys, setdefault methods
popped_value = my_dict.pop('location')  # Pops 'location' key
print("Popped Value:", popped_value)
```

```python
print(my_dict)
```

```python
key, value = my_dict.popitem()  # Pops last inserted key-value pair
print("Popitem:", key, value)
```

```python
new_dict = dict.fromkeys(['a', 'b', 'c', 'd'], 100)  # Creates new dict from keys with same value
print("Fromkeys:", new_dict)
```
